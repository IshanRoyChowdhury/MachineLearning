## 2. Haberman Dataset Introduction

Data Set Information (Source: UCI Machine learning repository https://archive.ics.uci.edu/ml/datasets/Haberman's+Survival):

The dataset contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.


Attribute Information:

1. Age of patient at time of operation (numerical)
2. Patient's year of operation (year - 1900, numerical)
3. Number of positive axillary nodes detected (numerical)
4. Survival status (class attribute)
-- 1 = the patient survived 5 years or longer
-- 2 = the patient died within 5 year


Relevant Papers:

Haberman, S. J. (1976). Generalized Residuals for Log-Linear Models, Proceedings of the 9th International Biometrics Conference, Boston, pp. 104-122.

Landwehr, J. M., Pregibon, D., and Shoemaker, A. C. (1984), Graphical Models for Assessing Logistic Regression Models (with discussion), Journal of the American Statistical Association 79: 61-83.

Lo, W.-D. (1993). Logistic Regression Trees, PhD thesis, Department of Statistics, University of Wisconsin, Madison, WI.

## Univariate Analysis 

```python
import numpy as np
people_survived = data.loc[data['status'] == 1];
people_expired = data.loc[data["status"] == 2];

plt.plot(people_survived['nodes'], np.zeros_like(people_survived['nodes']), 'x', label='survived');
plt.plot(people_expired['nodes'], np.zeros_like(people_expired['nodes']),'o', label='expired');
plt.xlabel('Number of nodes')
plt.legend()
plt.title('Univariate Analysis on Nodes')
plt.show()
```

Observations:

Let us take a look at the pdf or the histogram for 'Nodes'

## Histogram or Probability Density Function based Analysis

```python
#pdf for nodes, setting status as hue, so that we have a pdf for both the status
# status 1 means patient survived
# status 2 means patient died
sns.FacetGrid(data, hue="status", size=8) \
   .map(sns.distplot, "nodes") \
   .add_legend();
plt.title("PDF for nodes seperated by Status",size=20)
plt.show();
```

#### First breakthrough:

Finally some promising insight.

The two probability density functions overlap but can also be well distinguished as apparent from the plot above.
The pdf for 'Surviving patients' have a small spread i.e standard deviation is low, whereas the spread for subjects who 'Died' have a much larger spread relatively and this high variance in the second pdf explains why it has been so difficult to seperate the two, using just scatter plots (as shown in pair plots above). 

In addition in this plots is even more significant as the dataset is imbalanced, 73% dataset comprises of surviving patients and despite that the stdandard deviation of "surving patient's nodes" distribution has very low standard deviation. So there is some significant correlation in between the number of nodes and the final fate of the patients.


At 4 nodes we see the pdf of subjects who died exceed the pdf of surviving subjects. This is a strong candidate for thresholding. 

First draft of model:

if(data.Nodes > 4):
   status = 2
else:
   status = 1
   
Let us test this out!


```python
def Model1(num_nodes):
    '''
    Model 1 classifies whether a patient survives or dies solely based on the number of nodes they had at the time
    of operation.
    
    @param_in: number of nodes
    @return : void
    
    '''
    pred_true = 0
    pred_false = 0
    idx_node = 0
    idx_status = 0

    for i in range(len(data)):
        row = data.iloc[i]
        
        if(row['nodes'] > num_nodes):
            #our model says Subject dies
            if(row['status'] == 2):
                pred_true +=1
            else:
                pred_false += 1
            
        else:
            #our model says subject lives
            if(row['status'] == 1):
                pred_true += 1
            else:
                pred_false += 1
        
    accuracy = pred_true/(pred_true + pred_false)
    print("Accuracy with {} nodes is {}".format(num_nodes,accuracy))
    
Model1(4)
Model1(5)
```

Observations:
1. Our model is able to classify whether a patient would have survived or died with 74.183% accuracy

2. 74.183 % accuracy is not bad, considering that I have just analysed based on nodes.
   This reaffirms our previous deduction on 'nodes' being one of the better attributes of this dataset for getting     
   insight. 

Let us now have a look at the pdf of other features.


```python
#pdf for age, similar to 'nodes' we will be using 'status' as hue
sns.FacetGrid(data, hue="status", size=6) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.title("PDF for age seperated by Status")
plt.show();
```

Age can be the second best feature. From the data we can see that beyond 40-42 year old patients the likelihood of patient dying is higher. But the chances of surviving exceeds again at around 58 years. And then till age of 75, it is hard to tell whether the patient will survive or die. Beyond 75 again we can assume that the likelihood of them dying exceeds their chances of living.

We can incorporate the above observations in our model, but it seems as if we are thentrying to overfit our model and that is bound to fail if we test it on another dataset. 

I think irrespective of the nodes count , we can safely infer the following:
1. age 30-42 patient survives
2. age 42-55 patient dies
3. age 56-78 grey area
4. age 78+ patient dies.

Next let us test the above observations and see how accurately can we classify using the 'age' feature.

```python
#writing a model just with the above conclusions regarding age
pred_true = 0
pred_false = 0
idx_node = 0
idx_status = 0

for i in range(len(data)):
    row = data.iloc[i]
    
    if (row['age'] <= 42):
        
        #our model says Subject lives
        if(row['status'] == 1):
            pred_true +=1
        else:
            pred_false += 1
            
    elif(row['age'] > 42 and row['age'] <= 55):
        
        #our model says subject dies
        if(row['status'] == 2):
            pred_true += 1
        else:
            pred_false += 1
            
    elif(row['age'] > 55 and row['age'] < 77):
        #grey area so also using nodes to predict
        if(row['nodes'] > 4):
            #patient dies
            if(row['status'] == 2):
                pred_true +=1
            else:
                pred_false += 1
        else:
            #patient lives
            if(row['status'] == 1):
                pred_true += 1
            else:
                pred_false +=1
    
    else:
        #our model says subject dies
        if(row['status'] == 2):
            pred_true += 1
        else:
            pred_false += 1
                
print("Total count",pred_true + pred_false)
print("Number of patients whose fate were not predicted",data.shape[0]-(pred_true + pred_false))
accuracy = pred_true/(pred_true + pred_false)
print("Accuracy of model with just 'age' feature is {}%".format(accuracy*100))
```

So our Model based primarily on 'Age' did fairly well with 60% accuracy. However, we had to use Nodes for deciding in the gray area of 56-76 years old subjects.

Our deduction on 'age' being a good candidate for second best feature is supported with the above results.
We now need to find the correct way to fuse these two features, so that our model gets more accurate.

## Cumulative Distribution Function


```python
#plotting the cdf for nodes with higher bin size to better dinstinguish
counts, bin_edges = np.histogram(data['nodes'], bins=50, 
                                 density = True)
pdf = counts/(sum(counts))
print("pdf:\n",pdf);
print("\nBin Edges:\n",bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
print("\ncdf:\n",cdf)
plt.plot(bin_edges[1:],pdf,label='pdf')
plt.plot(bin_edges[1:], cdf, label = 'cdf')
plt.xlabel('Nodes')
plt.ylabel('Density')
plt.legend()
plt.title('CDF for NODES')
plt.show();
```

Observations:

1. 57% of patients have less than or equal to 1.04 nodes
2. From the cdf we can infer that 75% of subjects have less than 4 nodes.
3. At around bin size 15,we have 93% of the data points, implying that only 7% subjects had more than 16 lymph nodes.

This gets interesting! With CDF we get numerical insights which we did not have till now.
Next we look at nodes distribution seperately in form of patients who survived and those who died.


```python
#Bin size lower to get smoother curves as multiple plots

#Patients with Status 1: survived

counts, bin_edges = np.histogram(people_survived['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print("\nPatients who Survived:\n")
print("pdf:\n",pdf);
print("Bin Edges:\n",bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label = 'pdf_survived')
plt.plot(bin_edges[1:], cdf, label = 'cdf_survived')

print("CDF:\n", cdf)

# Patients with status 2: died
counts, bin_edges = np.histogram(people_expired['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print("\nPatients who died:\n")
print("pdf:\n",pdf);
print("Bin Edges:\n",bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label='pdf_died')
plt.plot(bin_edges[1:], cdf, label = 'cdf_died')
print("CDF:\n",cdf)
plt.legend()
plt.xlabel('Nodes')
plt.ylabel('Density')
plt.title('CDF of nodes for patients of both Status')
plt.show();
```

Observations:
    1. 91.6 % of patients who survived have less than or equal to 4.6 nodes
    2. 56% of patients who died also had 0-5 nodes, however that's from 26% of dataset.
    3. 60 % of patients who died as well had lymp nodes less than 4.16 
       (this was observed by increasing the bin size to 25)
    4. So with our threshold of 4 nodes, we will misclassify approximately 60% of patients who died,
       however 60 % of 26% of dataset is around 15.6%.
       We had a misclassification of 26% with our model 1, so the rest 10.4% is coming from 8-9% of patients who
       survived and had more than 4 lymph nodes.
